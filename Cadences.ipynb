```python
i_SMS = []
i_Email = []
emails = []
SMS = []

# Extract data from 'Upcoming' section
for time in c["Upcoming"]:
    if time["TouchType"] == "Email":
        i_Email.append(time["Scheduled_Time"])
    else:
        i_SMS.append(time["Scheduled_Time"])

# Extract data from 'Emails' and 'SMS' sections
for email in c["Emails"]:
    emails.append(email['Created_Time'])
for mail in c["SMS"]:
    SMS.append(mail["Created_Time"])

def process_data(data):
    # Process data to get unique dates and counts
    emails = [i[5:10] for i in data]
    unique_dates = sorted(set(emails))
    date_counts = Counter(emails)
    sorted_dates = sorted(date_counts.keys())
    counts = [date_counts[date] for date in sorted_dates]
    return unique_dates, sorted_dates, counts

# Process data from i_SMS
unique_dates1, sorted_dates1, counts1 = process_data(i_SMS)
# Process data from i_Email
unique_dates2, sorted_dates2, counts2 = process_data(i_Email)
# Get all unique dates
all_dates = set(sorted_dates1).union(sorted_dates2)

# Initialize counters for date counts
date_counts1 = Counter({date: 0 for date in all_dates})
date_counts2 = Counter({date: 0 for date in all_dates})

# Assign counts to the corresponding dates
for date in sorted_dates1:
    date_counts1[date] = counts1.pop(0)

for date in sorted_dates2:
    date_counts2[date] = counts2.pop(0)

# Sort dates and counts based on sorted dates
sorted_dates = sorted(all_dates)
counts1 = [date_counts1[date] for date in sorted_dates]
counts2 = [date_counts2[date] for date in sorted_dates]


plt.figure(figsize=(18, 10))
plt.bar(sorted_dates, counts2, label='Email')
plt.bar(sorted_dates, counts1, bottom=counts2, label='SMS', color="orange")
plt.xlabel('Dates')
plt.ylabel('Count')
plt.title('Upcoming Touches')
plt.legend(loc="upper right")
plt.xticks(rotation=45)
plt.savefig('Upcoming.png')
for i in range(len(sorted_dates)):
    plt.text(i, (counts1[i]+counts2[i])/2, counts2[i]+counts1[i], ha = 'center', color = "black")
plt.tick_params(left = False, right = False, labelleft = False)
plt.show()


# Process data from SMS
unique_dates3, sorted_dates3, counts3 = process_data(SMS)
# Process data from emails
unique_dates4, sorted_dates4, counts4 = process_data(emails)
# Get all unique dates
all_dates1 = set(sorted_dates3).union(sorted_dates4)
# Initialize counters for date counts
date_counts3 = Counter({date: 0 for date in all_dates1})
date_counts4 = Counter({date: 0 for date in all_dates1})

# Assign counts to the corresponding dates
for date in sorted_dates3:
    date_counts3[date] = counts3.pop(0)

for date in sorted_dates4:
    date_counts4[date] = counts4.pop(0)

# Sort dates and counts based on sorted dates
sorted_dates1 = sorted(all_dates1)
counts3 = [date_counts3[date] for date in sorted_dates1]
counts4 = [date_counts4[date] for date in sorted_dates1]

plt.bar(sorted_dates1, counts4, label='Email')
plt.bar(sorted_dates1, counts3, bottom=counts4, label='SMS', color="orange")
plt.xlabel('Dates')
plt.ylabel('Total Sent')
plt.title('Sent Emails and SMS per Cadence')
plt.legend(loc="upper right")
plt.xticks(rotation=45)
for i in range(len(sorted_dates1)):
    plt.text(i, (counts3[i]+counts4[i])/2, counts3[i]+counts4[i], ha = 'center', color = "black")
plt.tick_params(left = False, right = False, labelleft = False)

plt.savefig('Cadences.png')
plt.show()
```

```python
opens = []
clicks = []

for email in c["Emails"]:
    opens.append(email["Opens"])
    clicks.append(email["Clicks"])

modified_opens = [0 if x is None else 1 for x in opens]
modified_clicks = [0 if x is None else 1 for x in clicks]

average_opens = (sum(modified_opens) / len(modified_opens))*100
average_clicks = (sum(modified_clicks) / len(modified_clicks))*100
opens_to_clicks = (average_clicks/average_opens)*100

print(f"Open Rate: {average_opens}%")
print(f"Click Rate: {average_clicks}%")
print(f"Click rate to Open rate: {opens_to_clicks}%")
```

```python
open = []
click = []
total_SMS = []
total_Email = []
dates = []
unique_list = []

for x in c["Emails"]:
    if x["deliver_status"] == "delivered":
        dates.append(x["Created_Time"])
dates = [i[5:10] for i in dates]

for x in dates:
    if x not in unique_list:
        unique_list.append(x)

sorted_date = sorted(unique_list)

sent_data = []

for date in sorted_date:
    opens = 0
    clicks = 0
    email_count = 0
    sms_count = 0

    for email in c["Emails"]:
        if email["Created_Time"][5:10] == date and email["deliver_status"] == "delivered":
            if email["Opens"] is not None:
                opens += email["Opens"]
            if email["Clicks"] is not None:
                clicks += email["Clicks"]
            email_count += 1

    for sms in c["SMS"]:
        if sms["Created_Time"][5:10] == date and sms["deliver_status"] == "delivered":
            sms_count += 1

    sent_data.append({
        "Date": date,
        "Emails Sent": email_count,
        "Emails Opened": opens,
        "Emails Clicked": clicks,
        "SMS Sent": sms_count
    })

seeds_counts = {}
Emails_Scheduled_counts = {}
SMS_Scheduled_counts = {}

for time in c["Upcoming"]:
    if time is not None:
        scheduled_time = time["Scheduled_Time"][5:10]
        if scheduled_time not in seeds_counts:
            seeds_counts[scheduled_time] = 0
        if scheduled_time not in Emails_Scheduled_counts:
            Emails_Scheduled_counts[scheduled_time] = 0
        if scheduled_time not in SMS_Scheduled_counts:
            SMS_Scheduled_counts[scheduled_time] = 0

        if time["TouchType"] == "Email":
            Emails_Scheduled_counts[scheduled_time] += 1
        else:
            SMS_Scheduled_counts[scheduled_time] += 1

        seeds_counts[scheduled_time] += 1

upcoming_data = []

for scheduled_time in sorted(seeds_counts.keys()):
    upcoming_data.append({
        "Date": scheduled_time,
        "Seeds": seeds_counts[scheduled_time],
        "Emails Scheduled": Emails_Scheduled_counts[scheduled_time],
        "SMS Scheduled": SMS_Scheduled_counts[scheduled_time]
    })

data = {
    "Sent": sent_data,
    "Upcoming": upcoming_data
}

print(data)
```
